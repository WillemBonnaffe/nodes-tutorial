```python
####################
## INITIALIZATION ##
####################

## Import required libraries
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torchdiffeq import odeint
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns

## Set random seed for reproducibility
torch.manual_seed(0)

#
###
```

```python
################################
## LOADING AND EXPLORING DATA ##
################################

## Goal:
## Load and preprocess the Hudson Bay Companyâs hare-lynx dataset, visualize basic trends, and normalize data for model training.

## Load data function
## Goal:
## Load the hare-lynx dataset from a CSV file and preprocess for NODE training.
## Inputs:
## file_path (str): Path to the CSV file containing hare-lynx time series data.
## Outputs:
## data_tensor (torch.Tensor): Normalized hare-lynx population data as a PyTorch tensor.
def load_and_preprocess_data(file_path):

    ## Load the data
    data = np.array(pd.read_csv(file_path))
    data = data[:,1:3] # Drop time step

    ## Subset time steps
    # data = data[0:30]

    ## Normalize data to stabilize training
    data_normalized = np.copy(data)
    for i in range(data.shape[1]):
        data_log = np.log(data[:,i])
        data_normalized[:,i] = (data_log - data_log.mean()) / data_log.std()
    
    ## Convert data to PyTorch tensor
    data_tensor = torch.tensor(data_normalized, dtype=torch.float32)
    
    return data_tensor

## Load and visualize the data
data_tensor = load_and_preprocess_data('data_hare_lynx.csv')
num_time_steps = data_tensor.size()[0]
data_tensor = data_tensor.float()

## Plot the data to observe trends in hare and lynx populations
plt.figure(figsize=(14,6))
plt.plot(data_tensor[:, 0], '--o', label='Hare Population', color='royalblue')
plt.plot(data_tensor[:, 1], '--o', label='Lynx Population', color='salmon')
plt.xlabel('Year')
plt.ylabel('Normalized Population')
plt.legend()
plt.title('Hare-Lynx Population Time Series')
sns.despine()
plt.show()

#
###
```

```python
##############################################
## 3D VISUALIZATION OF HARE-LYNX POPULATION ##
##############################################

## Goal:
## Visualize the hare and lynx population densities over time in a 3D plot with time on the z-axis.
## The line color changes according to the time axis, with early time points in blue and later time points in red.

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

## Prepare data for 3D plot
hare_density = data_tensor[:, 0].numpy()  # Hare population density
lynx_density = data_tensor[:, 1].numpy()  # Lynx population density
time_steps = np.arange(num_time_steps)    # Time steps

## Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

## Normalize time steps for color mapping
norm = plt.Normalize(time_steps.min(), time_steps.max())
colors = cm.plasma(norm(time_steps))  # Use 'plasma' colormap for gradient from blue to red

## Plot each point in the time series with color corresponding to time
ax.plot(hare_density, lynx_density, time_steps, '--', color='purple', alpha=0.5)
ax.scatter(hare_density, lynx_density, time_steps, color=colors, marker='o', s=10, cmap='jet')

## Set axis labels and title
ax.set_xlabel('Hare Density')
ax.set_ylabel('Lynx Density')
ax.set_zlabel('Time')
ax.set_title('3D Hare-Lynx Population Dynamics Over Time (Color-coded by Time)')

## Display color bar for reference
mappable = cm.ScalarMappable(cmap=cm.plasma, norm=norm)
mappable.set_array(time_steps)
fig.colorbar(mappable, ax=ax, label="Time")

plt.show()

#
###
```

# Latent variable NODE

```python
##########################
## LATENT VARIABLE NODE ##
##########################

## Goal:
## Introduce a latent variable to the NODE model to explore if an additional unobserved factor can improve population predictions.
```

```python
##################
## DEFINE MODEL ##
##################

## Extended NODE model class with latent variable
## Goal:
## Define a neural network with a latent variable representing an unobserved factor affecting hare-lynx dynamics.
## Inputs:
## N/A
## Outputs:
## dX/dt (torch.Tensor): Derivative of hare, lynx, and latent variables.
class PredatorPreyODE_Latent(nn.Module):

    def __init__(self):
        super(PredatorPreyODE_Latent, self).__init__()

        ## Define a network with additional latent variable in output
        self.net = nn.Sequential(
            nn.Linear(3, 50),   # Input layer with two populations and one latent variable
            nn.Tanh(),
            nn.Linear(50, 3)    # Output layer with three outputs: dHare/dt, dLynx/dt, dLatent/dt
        )

    def forward(self, t, x):
        return self.net(x)


#
###
```

```python
###############
## FUNCTIONS ##
###############

## Sequential training loop with dynamic weighted least squares
## training_loop_weighted function
## Goal:
## Train the NODE model with a latent variable over a specified number of epochs, using weighted least squares to emphasize early time points.
## Inputs:
## model (nn.Module): The NODE model with latent variable to be trained.
## optimizer (torch.optim.Optimizer): Optimizer for adjusting model weights.
## data (torch.Tensor): Time series data of hare and lynx populations.
## epochs (int): Number of epochs to run the training loop.
## reg_lambda (float): Regularization parameter for L2 penalty on the latent state.
## weight_decay_schedule (function): A function that defines how the weights decay over epochs.
## Outputs:
## None
def training_loop_weighted(model, optimizer, data, epochs, reg_lambda=0.001, decay_rate=0.1):

    ## Time points
    time_points = torch.linspace(0, data.shape[0], steps=data.shape[0])  # Define time points for integration
    
    ## Initial weighting for each time step, emphasizing early points
    weights = torch.exp(-decay_rate * torch.linspace(0.0, data.shape[0], steps=data.shape[0])) # Initial weights decreasing over time
    # weights = torch.linspace(1.0, 0.00001, steps=data.shape[0]) ** 2 # Initial weights decreasing over time

    ## Softmax on weights
    smax = nn.Softmax(dim=0)

    for epoch in range(epochs):

        ## Update weights dynamically over time (progressively flattening the weights)
        # weight_decay_factor = 1 - (epoch / epochs)
        # adjusted_weights = smax(weights * weight_decay_factor + (1 - weight_decay_factor) * torch.ones_like(weights))
        adjusted_weights = weights / weights.sum() # Initial weights decreasing over time

        ## Initialize the initial conditions with zero latent variable for each epoch
        initial_conditions = torch.cat([data[0], torch.tensor([0.0])])  # (hare, lynx, latent)

        ## Set gradients to zero before each pass
        optimizer.zero_grad()

        ## Forward pass: use ODE solver to predict populations over time
        predicted_population = odeint(model, initial_conditions, time_points)

        ## Compute weighted least squares loss (only for hare and lynx predictions)
        squared_errors = (predicted_population[:, :2] - data) ** 2
        weighted_mse_loss = torch.sum(adjusted_weights.view(-1, 1) * squared_errors)

        ## Compute L2 regularization on the latent state
        latent_states = predicted_population[:, 2]  # Extract the latent variable over all time steps
        latent_reg_loss = reg_lambda * torch.sum(latent_states ** 2)  # L2 penalty on latent state

        ## Total loss is weighted MSE loss + regularization loss
        total_loss = weighted_mse_loss + latent_reg_loss

        ## Backward pass: compute gradients and update model parameters
        total_loss.backward()
        optimizer.step()

        ## Log training progress every 100 epochs
        if epoch % 100 == 0:
            print(f"Epoch {epoch}/{epochs}, Weighted MSE Loss: {weighted_mse_loss.item():.4f}, "
                  f"Latent Reg Loss: {latent_reg_loss.item():.4f}, Total Loss: {total_loss.item():.4f}")

#
###
```

```python
## evaluate_model_latent function
## Goal:
## Evaluate the latent variable NODE model by comparing predictions of hare and lynx populations to actual data.
## Inputs:
## model (nn.Module): The trained NODE model with latent variable.
## data (torch.Tensor): Time series data of hare and lynx populations.
## Outputs:
## final_loss (float): Final MSE loss after model training.
def evaluate_model_latent(model, data):

    ## Time points
    time_points = torch.linspace(0, data.shape[0], steps=data.shape[0])  # Define time points for integration
    
    ## Set initial conditions including latent variable
    initial_conditions = torch.cat([data[0], torch.tensor([1.5])])

    ## Perform forward pass to get model predictions
    predicted_population = odeint(model, initial_conditions, time_points)

    ## Calculate final MSE loss on the dataset
    final_loss = mse_loss(predicted_population[:, :2], data).item()
    print(f"Final Loss (with Latent Variable): {final_loss:.4f}")

    ## Visualize the results (ignore latent variable in plot)
    plt.figure(figsize=(14,6))
    plt.plot(data[:, 0], '--o', label="True Hare Population", color='royalblue')
    plt.plot(data[:, 1], '--o', label="True Lynx Population", color='salmon')
    plt.plot(predicted_population[:, 0].detach(), label="Predicted Hare with Latent Variable", color='royalblue')
    plt.plot(predicted_population[:, 1].detach(), label="Predicted Lynx with Latent Variable", color='salmon')
    plt.xlabel('Year')
    plt.ylabel('Normalized Population')
    plt.legend()
    plt.title('True vs Predicted Hare-Lynx Dynamics with Latent Variable')
    plt.show()

    return final_loss

#
###
```

```python
####################
## INITIATE MODEL ##
####################

## Initialize the extended model
predator_prey_ode_latent = PredatorPreyODE_Latent()

## Define initial conditions for hare, lynx, and latent variable
initial_conditions_latent = torch.cat([data_tensor[0], torch.tensor([0.0])])  # Adding initial latent value of 0

## Define training parameters
mse_loss = nn.MSELoss()
optimizer_latent = optim.Adam(predator_prey_ode_latent.parameters(), lr=0.001)
num_epochs = 1000

#
###
```

```python
#####################################
## RUNNING TRAINING AND EVALUATION ##
#####################################

## Goal:
## Execute the training loop and evaluate the NODE model with latent variable on the hare-lynx dataset.

## Run training loop for model with latent variable
training_loop_weighted(model=predator_prey_ode_latent, optimizer=optimizer_latent, data=data_tensor, epochs=num_epochs, reg_lambda=0.0001, decay_rate=0.1)

## Evaluate model after training
final_loss_latent = evaluate_model_latent(predator_prey_ode_latent, data_tensor)

#
###
```

```python
#######################################
## ANALYSIS: PREDICTED DYNAMICS GRID ##
#######################################

## Goal:
## Analyze the NODE model's dynamics by evaluating it over a grid of hare and lynx densities. 
## Display two heatmaps showing the predicted dynamics of hare and lynx populations.

import numpy as np
import matplotlib.pyplot as plt

## dynamics_grid_analysis function
## Goal:
## Evaluate NODE model dynamics over a grid of hare and lynx densities, and plot heatmaps of predicted dynamics.
## Inputs:
## model (nn.Module): Trained NODE model (with or without latent variable).
## hare_range (tuple of float): Range of hare densities to evaluate (min, max).
## lynx_range (tuple of float): Range of lynx densities to evaluate (min, max).
## latent_value (float): Value of the latent variable.
## grid_size (int): Number of points along each axis in the grid (default=50).
## Outputs:
## None
def dynamics_grid_analysis(model, hare_range, lynx_range, latent_value, grid_size=50):
    
    ## Create mesh grid for hare and lynx densities
    hare_values = np.linspace(hare_range[0], hare_range[1], grid_size)
    lynx_values = np.linspace(lynx_range[0], lynx_range[1], grid_size)
    hare_grid, lynx_grid = np.meshgrid(hare_values, lynx_values)

    ## Initialize arrays to store dynamics predictions for each grid point
    hare_dynamics = np.zeros((grid_size, grid_size))
    lynx_dynamics = np.zeros((grid_size, grid_size))
    
    ## Evaluate dynamics across the grid
    for i in range(grid_size):
        for j in range(grid_size):
            ## Prepare hare and lynx populations as a torch tensor for model input
            hare_density = hare_values[i]
            lynx_density = lynx_values[j]
            input_tensor = torch.tensor([hare_density, lynx_density, latent_value], dtype=torch.float32)

            ## Get dynamics prediction (dHare/dt, dLynx/dt) from the model
            dynamics = model(0, input_tensor)
            
            ## Store the predicted dynamics in the arrays
            hare_dynamics[i, j] = dynamics[0].item() # dHare/dt
            lynx_dynamics[i, j] = dynamics[1].item() # dLynx/dt

    ## Plot heatmap for hare population dynamics
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(hare_dynamics, extent=(hare_range[0], hare_range[1], lynx_range[0], lynx_range[1]), origin='lower', cmap='jet')
    plt.colorbar(label='dHare/dt')
    plt.xlabel("Hare Density")
    plt.ylabel("Lynx Density")
    plt.title("Predicted Dynamics of Hare Population (dHare/dt)")

    ## Plot heatmap for lynx population dynamics
    plt.subplot(1, 2, 2)
    plt.imshow(lynx_dynamics, extent=(hare_range[0], hare_range[1], lynx_range[0], lynx_range[1]), origin='lower', cmap='jet')
    plt.colorbar(label='dLynx/dt')
    plt.xlabel("Hare Density")
    plt.ylabel("Lynx Density")
    plt.title("Predicted Dynamics of Lynx Population (dLynx/dt)")

    plt.tight_layout()
    plt.show()


## Running dynamics grid analysis
## Define hare and lynx density ranges based on the observed data
hare_min, hare_max = data_tensor[:, 0].min().item(), data_tensor[:, 0].max().item()
lynx_min, lynx_max = data_tensor[:, 1].min().item(), data_tensor[:, 1].max().item()

## Execute analysis
dynamics_grid_analysis(predator_prey_ode_latent, hare_range=(hare_min, hare_max), lynx_range=(lynx_min, lynx_max), latent_value=-1.0)
dynamics_grid_analysis(predator_prey_ode_latent, hare_range=(hare_min, hare_max), lynx_range=(lynx_min, lynx_max), latent_value=0.0)
dynamics_grid_analysis(predator_prey_ode_latent, hare_range=(hare_min, hare_max), lynx_range=(lynx_min, lynx_max), latent_value=1.0)

#
###
```
